```python
import numpy as np
import pandas as pd
from itertools import combinations, product
```

```python
num_hidden_people_features = 6
num_hidden_movie_features = 5

num_users = 50
num_movies = 120 
seen_pct = 0.7
```

```python
users = np.arange(num_users)
np.random.seed(42)
user_features = np.random.randn(num_users, num_hidden_people_features)
```

```python
movies = np.arange(num_movies)
np.random.seed(42)
movie_features = np.random.randn(num_movies, num_hidden_movie_features)
movie_min = movie_features.min()
movie_max = movie_features.max()
movie_features = ((movie_features - movie_min) / (movie_max - movie_min)).round(1)
```

```python
def utility(user_features, movie_features):
    return user_features[1:] @ movie_features + user_features[0]
```

```python
df = pd.DataFrame(index=users, columns=movies)
```

```python
prod = list(product(users.tolist(), movies.tolist()))
```

```python
np.random.seed(42)
np.random.shuffle(prod)
```

```python
# for i in range(int(num_users * num_movies * seen_pct)):
for i in range(len(prod)):
    u, m = prod[i]
    df.loc[u, m] = utility(user_features[u], movie_features[m])
```

```python
scale_min = df.astype(float).min().min()
scale_max = df.astype(float).max().max()
```

```python
df = ((df.astype(float) - scale_min) / (scale_max - scale_min) * 5.).round()
```

```python
df = df.T
```
