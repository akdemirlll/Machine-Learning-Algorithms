```python
_ =df.reset_index().melt(id_vars="movie_id") # pivot table ı geri almak için kullanılır
```

```python
# y= X @ theta
# elimizde X ve theta yok
```

```python
X = np.random.randn(120,6)
X[:,0] = 1.0
theta = np.random.randn(6,50 )
```

```python
y_pred = X @ theta
```

```python
df- y_pred
```

```python
# error =(df- y_pred)**2
error = y_pred - df # türevinden gelen değer
```

## Backpropagation

```python
error @ theta.T 
```

```python
(error.fillna(0.0)) @ theta.T
```

```python
lr = 0.001

for i in range(1000):
    y_pred = X @ theta
    error = y_pred - df

    X_new = X - lr * (error.fillna(0.0)) @ theta.T
    theta = theta - lr * (X.T @ error.fillna(0.0)) # (6,50)

    X = X_new

print(np.nanmean((error**2).values))
```
