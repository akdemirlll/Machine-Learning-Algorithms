## Hata fonksiyonu

```python
def hinge_loss_deneme(y, ypred):
    if y == 1:
        if ypred >= 1:
            return 0
        else:
            return y - ypred
    else:
        if ypred <= -1:
            return 0
        else:
            return ypred - y
```

## SVM

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=100, centers=2, n_features=2, cluster_std=2, random_state=42)
```

```python
plt.scatter(*X.T, c=y)
```

```python
def decision_func(X, b0, b1, b2):
    return b0 + b1 * X.T[0] + b2 * X.T[1]
```

```python
t = decision_func(X, 0,-1,1)
```

```python
plt.scatter(*X.T, c=t)
plt.colorbar()
```

```python
def SVM(X):
    
    ypred = X.copy()
    
    ypred[ypred > 0] = 1
    ypred[ypred <= 0] = 0
    
    return ypred.astype(int)
```

```python
def hinge_loss(y, decision):
    
    y = np.array(y).copy()
    decision = np.array(decision)
    
    y[y == 0] = -1
        
    loss = 1 - y * decision
    loss[loss < 0] = 0
    
    return loss
```

## Y = 1 için

```python
def hinge_loss(y, ypred):
    
    y = np.array(y).copy()
    decision = np.array(ypred)
    
    y[y == 0] = -1
        
    loss = 1 - y * ypred
    loss[loss < 0] = 0
    
    return loss
```

```python
decision = np.linspace(-10,10,500)
```

```python
L = hinge_loss(1,decision)
```

```python
L[L > 0] = -1
```

```python
plt.plot(decision, L)
```

## SVM Optimization

```python
from warnings import warn
```

```python
class SVM_optima:

    def __init__(self, C=1, lr=0.01, tol=10**-2, max_iter=1000):
        self.w = None
        self.b = None
        self.lr=lr
        self.tol=tol
        self.max_iter=max_iter
        self.C=C

    def __repr__(self):
        return f"SVM()"

    def gradient(self, X, y):
        
        y = y.copy()
        y[y == 0] = -1

        C = self.C #pratik olması için
        w = np.random.randn(X.shape[1], 1)
        b = 0

        for iter_ in range(self.max_iter):
            ypred = X @ w + b  # feed_forward
                        
            miss_loss = hinge_loss(y, ypred.ravel()) 
            Loss = C * miss_loss.sum() + (w**2).sum() + b**2
            
            dmiss_loss = -y
            dmiss_loss[miss_loss == 0] = 0
            
            dLoss = dmiss_loss * C # dLoss / dypred
            
            # back_propogation
            dw = X.T @ (dLoss.reshape(-1,1)) + 2 * w 
            db = dLoss.sum() + 2 * b 
                        
            w = w - self.lr * dw
            b = b - self.lr * db

            is_converged = (np.abs(dw) < self.tol).all() and (np.abs(db) < self.tol) 
            if is_converged:
                break

        if not is_converged:
            warn('Model yakinsamadi. Azami iterasyon sayisini arttirmayi deneyiniz.')

        self.w = w
        self.b = b
        
    def tahmin_et(self, X):
        ypred = X @ self.w + self.b
        
        ypred[ypred > 0] = 1
        ypred[ypred <= 0] = 0

        return ypred.astype(int)
```

## Örnek çözüm

```python
X, y = make_blobs(n_samples=100, centers=2, n_features=2, cluster_std=3, random_state=42)

plt.scatter(*X.T, c=y)
plt.colorbar()
```

```python
svm = SVM_optima(lr = 1e-1)
```

```python
svm.gradient(X, y)
```

```python
ypred = svm.tahmin_et(X)
```

```python
plt.scatter(*X.T, c=ypred.ravel())
plt.colorbar()
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y, ypred)
```

## RBF Kernel

```python
X, y = make_blobs(n_samples=100, centers=2, n_features=2, cluster_std=3, random_state=42)
plt.scatter(*X.T, c=y)
plt.colorbar()
```

```python
X = X[y.argsort()]
y = y[y.argsort()]
```

```python
X.shape
```

```python
distance = (((X.reshape(100, 2, 1) - X.T.reshape(1, 2, 100))**2).sum(axis=1))**0.5
```

```python
distance.shape
```

```python
sigma = 1
```

```python
x_transformed = np.exp(-(distance**2 / (2 * sigma**2)) )
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (14,8))
sns.heatmap(x_transformed)
```

```python
svm_kernel = SVM_optima()
```

```python
svm_kernel.gradient(x_transformed, y)
```

```python
ypred = svm_kernel.tahmin_et(x_transformed)

plt.scatter(*X.T, c=ypred.ravel())
plt.colorbar()
```

```python
plt.scatter(*X.T, c=y)
plt.colorbar()
```

```python
metrics.accuracy_score(y, ypred)
```

```python
def kernel(X, y=None, sigma=1):
    if y is None:
        y = X
    
    distance = (((y.reshape(y.shape[0], y.shape[1], 1) - X.T.reshape(1, X.shape[1], X.shape[0]))**2).sum(axis=1))**0.5
    
    x_transformed = np.exp(-(distance**2 / (2 * sigma**2)) )
    
    return x_transformed    
```

```python
kernel(X) == x_transformed
```

```python
t = kernel(X, np.array([[5,0]]))
```

```python
svm_kernel.tahmin_et(t)
```
