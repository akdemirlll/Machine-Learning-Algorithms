```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X,y = make_blobs(n_samples=400, n_features=2, centers=6, random_state=40)
```

```python
plt.scatter(*X.T, c="gray")
```

```python
from sklearn.cluster import KMeans
```

# WCSS: Within Cluster Sum of Squares. inertia_

```python
num_clusters = range(1,15)
inertias= []
for k in num_clusters:
    model= KMeans(n_clusters=k)
    model.fit(X)
    inertias.append(model.inertia_)
```

```python
inertias
```

```python
plt.figure(figsize=(12,8))
plt.plot(num_clusters, inertias)
```

```python
model.predict(X)
```

```python
model.cluster_centers_[5]
```

```python
cluster_center_replace= model.cluster_centers_[model.predict(X)]
```

```python
((X-cluster_center_replace)**2).sum()
```

```python
model.inertia_
```

```python
from ipywidgets import interact
```

```python
@interact
def means(k=(1,15,1)):
    k=k
    model=KMeans(n_clusters=k)
    model.fit(X)
    print("inertia: ", model.inertia_)
    cluster= model.predict(X)
    _= plt.scatter(*X.T, c=cluster)
```

## Non-Uniform Clusters

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=300, noise=.1, random_state=42)
```

```python
plt.scatter(*X.T, color="gray") 
```

```python
num_clusters = range(2,15)
inertias= []
for k in num_clusters:
    model= KMeans(n_clusters=k)
    model.fit(X)
    inertias.append(model.inertia_)
```

```python
plt.figure(figsize=(14,8))  # evet KMeans başarısız
plt.plot(num_clusters, inertias)
```

```python
@interact
def means(k=(1,15,1)):    # evet KMeans başarısız
    k=k
    model=KMeans(n_clusters=k)
    model.fit(X)
    print("inertia: ", model.inertia_)
    cluster= model.predict(X)
    _= plt.scatter(*X.T, c=cluster)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(linkage="single")
```

```python
cluster = model.fit_predict(X)
```

```python
plt.scatter(*X.T, c=cluster)
```

```python
cluster
```

```python
cluster = model.fit_predict(X)
plt.scatter(*X.T, c=cluster)
```

# Gaussian Mixture 

```python
a = np.random.randn(200,2)
```

```python
a = a @ [[1,0.8],[0.8,1]] 
```

```python
plt.scatter(*a.T)
```

```python
b = np.random.randn(200,2) @ [[1,-0.8],[-0.8,1]] 
```

```python
plt.scatter(*a.T)
plt.scatter(*b.T)
```

```python
from sklearn.mixture import GaussianMixture 
```

```python
cluster = GaussianMixture(n_components=2)
```

```python
x = np.concatenate([a,b + 4.78])
```

```python
plt.scatter(*x.T, c=cluster.fit_predict(x))
```

```python
cluster.predict_proba(a)[:,1].argmin()
```
