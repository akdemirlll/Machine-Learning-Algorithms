### learning rate

```python
lr = 0.01
```

## batch_gradient_descent

```python
def batch_gradient_descent(X, y):
    cost_iter = []
    weights = np.random.randn(n, 1)
    
    for _ in range(100):
        y_pred = X @ weights
        costs = np.mean((y_pred-y)**2)
        weights = weights - X.T @ (lr * (y_pred-y)) * (1 / X.shape[0]) 
        cost_iter.append(costs)
    
    return weights, cost_iter
```

```python
def stochastic_gradient_descent(X, y):
    weights = [np.random.randn(X.shape[1], 1)]
    index = np.arange(X.shape[0])
    np.random.shuffle(index)
    X = X[index]
    y = y[index]
    
    costs = []
    for i in range(1):
        for k in index:
            x_s = X[k:k+1]
            y_s = y[k:k+1]
            ypred_s = x_s@weights[-1]
            dL = ypred_s - y_s
            dw = x_s.T @ dL
            new_w = weights[-1] - 0.001 * dw / x_s.shape[0]
            weights.append(new_w)
            loss = ((ypred_s - y_s)**2).mean()
            costs.append(loss)
            
    return weights, costs
```

```python
weights, costs = stochastic_gradient_descent(X, y)
plt.figure()
plt.plot(*np.array(weights)[:,:,0].T, '*-', markersize=1)
plt.figure()
plt.plot(pd.Series(costs).rolling(5000).mean())
```

## mini_batch_gradient_descent

```python
np.random.seed(42)
M = 90000
n = 2
X = np.random.randn(M, n)
b = np.random.randn(n, 1)
e = np.random.randn(M, 1)
y = X@b + e
```

```python
def generate_batches(*args, batch_size=32):
    num_batches = int(np.ceil(len(args[0]) / 32))
    for i in range(num_batches):
        yield [arg[i:i+batch_size] for arg in args]
```

```python
def mini_batch_gradient_descent(X, y):
    b = 32
    
    weights = [np.random.randn(X.shape[1], 1)]
    index = np.arange(X.shape[0])
    
    costs = []
    
    for i in range(1):
        for x_s, y_s in generate_batches(X,y, batch_size = b):
            ypred_s = x_s@weights[-1]
            dL = ypred_s - y_s
            dw = x_s.T @ dL
            new_w = weights[-1] - 0.001 * dw / x_s.shape[0]
            weights.append(new_w)
            loss = ((ypred_s - y_s)**2).mean()
            costs.append(loss)
    
    return weights, costs
```

```python
weights, costs = mini_batch_gradient_descent(X, y)
plt.figure()
plt.plot(*np.array(weights)[:,:,0].T, '*-', markersize=1)
plt.figure()
plt.plot(pd.Series(costs).rolling(5).mean())
```
